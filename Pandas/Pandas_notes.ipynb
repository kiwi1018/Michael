## Series 級數
Series是一維標記的數組，能夠保存任何數據類型（整數，字符串，浮點數，Python對像等）。軸標籤統稱為索引。創建系列的基本方法是調用：
> s = pd.Series(data, index=index)  

- index若無填寫,預設為0,1,2,3

在這裡，data可以有很多不同的東西：
* Python字典
* ndarray
* scalar value

[參考資料:Intro to data structures](https://pandas.pydata.org/pandas-docs/stable/getting_started/dsintro.html)


### 例子1

```python
s1 = pd.Series(np.random.randn(5), index=['a', 'b', 'c', 'd', 'e'])
```

```python
s1
```

```python
d = {'a': 0., 'b': 1., 'c': 2.}
s2 = pd.Series(d, index=['b', 'c', 'd', 'a'])
```

```python
s2
```
